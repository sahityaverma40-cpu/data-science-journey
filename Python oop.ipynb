```python
class Employee:
    def __init__(self, name, age, department):
        self.name = name
        self.age = age
        self.department = department

    def display_info(self):
        return f"Name: {self.name}, Age: {self.age}, Department: {self.department}"


# Creating object
e = Employee("Sahitya", 21, "IT")

# Printing employee details
print(e.display_info())
```

```python
class Studant:
    def __init__(self,name,age,grade,section):
        self.name = name
        self.age = age
        self.grade = grade
        self.section = section
    def display_info(self):
        return f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}, Section: {self.section}"

s = Studant("Sahitya",20,"A","A")
print(s.display_info())
```
