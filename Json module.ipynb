```python
import json
data = {"name": "alice", "age": 30,"city":"New York"}
json_string = json.dumps(data)
print(json_string)
print(type(json_string))
```

```python
with open('data.json', 'w') as f:
    json.dump(data, f)
```

```python
json_data = '{"name": "Alice", "age": 25, "city": "New York"}'

python_obj = json.loads(json_data)
print(python_obj)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}
print(type(python_obj))  # <class 'dict'>
```

```python
with open("data.json", "r") as file:
    python_data = json.load(file)

print(python_data)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}
```

```python
class Car:
    def __init__(self, brand, model1):
        self.brand = brand
        self.model1 = model1

    def display_info(self):
        return f"{self.brand}, {self.model1}"

car1 = Car("Toyota", "Camry")
print(car1.display_info())
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private Attribute

    def get_balance(self):  # Getter
        return self.__balance

    def deposit(self, amount):  # Setter
        if amount > 0:
            self.__balance += amount

# Using Encapsulation
account = BankAccount(1000)
account.deposit(500)
print(account.get_balance())  # Output: 1500
```
